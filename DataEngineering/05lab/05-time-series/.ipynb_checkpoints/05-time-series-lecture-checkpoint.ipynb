## Date/time datatypes

The `pd.to_datetime` method tries to parse string representations of timestamps and convert them to `Timestamp` objects.

```python
pd.to_datetime('2018-01-15 3:45pm')
```

Note that some dates can be ambiguous – the date below can be interpreted as a US date (8 June 1952) or as an European date written with slashes. Pandas will default to the US notation there:

```python
pd.to_datetime('7/8/1952')
```

However, if the values make no sense, it will assume the European format (and give us an appropriate warning):

```python
pd.to_datetime('17/8/1952')
```

This behaviour can be made deterministic using `dayfirst`:

```python
pd.to_datetime('7/8/1952', dayfirst=True)
```

A list of dates can be used as well (you need to use `format='mixed'` if the strings are not uniform):

```python
pd.to_datetime(['2018-01-05', '7/8/1952', 'Oct 10, 1995'], format='mixed')
```

The format can also be specified explicitly:

```python
pd.to_datetime(['2/25/10', '8/6/17', '12/15/12'], format='%m/%d/%y')
```

## Frequencies

Pandas timeseries can have frequencies defined explicitly. They can be used to easily generate an index with regular timestamps:

```python
pd.date_range('1998-03-10', '1998-03-15', freq='D')
```

We can also specify the number of periods instead of the end date:

```python
pd.date_range('2004-09-20', periods=8, freq='H')
```

Or DataFrame's index doesn't have an explicit frequency yet:

```python
df.index
```

Let's take a subset of our dataset containing three irregularly-spaced dates:

```python
dates = pd.to_datetime(['2013-02-03', '2013-02-06', '2013-02-08'])
dfs = df.loc[dates, ['consumption']].copy()
dfs
```

```python
dfs.index
```

We can now use the `asfreq` method to fill in the missing dates. Note that the values are empty by default:

```python
dfsf = dfs.asfreq('D')
dfsf
```

We can use *forward fill* or *backwards fill* to fill these values; note that there's no interpolation going on at this point:

```python
dfsf['consumption_ffill'] = dfs.asfreq('D', method='ffill')
dfsf
```

## Resampling

The `asfreq` method only converts the DataFrame's index, keeping the original data. 

Data can also be resampled (either up or down) using the `resample` method.

### Downsampling

Downsampling means reducing the frequency of the data points. Instead of just dropping values, the values are aggregated, hence the need to specify an aggregation function. By calling `resample`, we obtain a *resample object*:

```python
df.resample('W')
```

It must be chained with an aggregation function, such as `mean`:

```python
dfw = df.resample('W').mean()
dfw
```

Let's plot the daily data and the weekly averages:

```python
start, end = '2017-01', '2017-06'
fig, ax = plt.subplots()
ax.plot(df.loc[start:end, 'solar'],
marker='.', linestyle='-', linewidth=0.5, label='Daily')
ax.plot(dfw.loc[start:end, 'solar'],
marker='o', markersize=8, linestyle='-', label='Weekly Mean Resample')
ax.set_ylabel('Solar Production (GWh)')
ax.legend();
```

When summing values, if all values are missing, the sum will be 0. This doesn't make sense for time series, so we can use the `min_count` argument to include only full months (at least 28 days seems like a reasonable condition):

```python
dfm = df.resample('M').sum(min_count=28)
dfm.head()
```

Let's plot the summarized montly data:

```python
fig, ax = plt.subplots()
ax.plot(dfm['consumption'], color='black', label='Consumption')
dfm[['wind', 'solar']].plot.area(ax=ax, linewidth=0)
ax.xaxis.set_major_locator(mdates.YearLocator())
ax.legend()
ax.set_ylabel('Monthly Total (GWh)');
```

Similarly, we can resample data to get annual sums:

```python
dfa = df.resample('A').sum(min_count=360)
dfa.head()
```

The “31 Dec” timestamps don't look very good. Let's use just the year for the index:

```python
dfa = dfa.set_index(dfa.index.year)
dfa.index.name = 'Year'
```

Finally, let's calculate the ratio of renewable production to consumption for each year to see if there is a trend:

```python
dfa['wind_solar_vs_consumption'] = dfa['wind_solar'] / dfa['consumption']
dfa.tail()
```

…and plot it:

```python
ax = dfa.loc[2012:, 'wind_solar_vs_consumption'].plot.bar(color='C0')
ax.set_ylabel('Fraction')
ax.set_ylim(0, 0.3)
ax.set_title('Wind + Solar Share of Annual Electricity Consumption')
plt.xticks(rotation=0);
```

### Upsampling

While downsampling required aggregation, upsampling requires some sort of interpolation.  

Of course, we expect the data to have daily repeatability (e.g. no solar production during the nights), but obviously interpolation will not reflect those.  Nevertheless, let's take a two-week period for experiments:

```python
dfs = df['2017-08-01':'2017-08-14'][['consumption']].copy()
dfs.plot();
```

Let's resample the data to hourly periods. Instead of an aggregation function, we will now supply an interpolator. The default one is linear interpolation:

```python
dfsh = dfs.resample('H').interpolate()
dfsh
```

Fill functions can also be used to “interpolate” the data:

```python
dfsh['ffill'] = dfs.resample('H').ffill()
dfsh['bfill'] = dfs.resample('H').bfill()
dfsh['nearest'] = dfs.resample('H').nearest()
```

```python
dfsh.plot()
```

More advanced interpolation functions are available from Numpy:

```python
dfsh = dfs.resample('H').interpolate()
dfsh['poly2'] = dfs.resample('H').interpolate('polynomial', order=2)
dfsh['poly3'] = dfs.resample('H').interpolate('polynomial', order=3)
dfsh['poly5'] = dfs.resample('H').interpolate('polynomial', order=5)
dfsh['spline'] = dfs.resample('H').interpolate('spline', order=2)
dfsh.plot()
```

## Rolling windows

Often, we want to smooth timeseries data to notice long-term trends. Calculating an average of adjacent periods (e.g. week-by-week) is not the best solution here. Instead, we use rolling windows. For example, if we use a 7-day centered rolling window, it means that for each day, an average will be calculated, including a period spanning three days back and three days forward:

```python
df7d = df.rolling(7, center=True).mean()
df7d.loc['2015'].head()
```

When we plot the data, we see that the rolling average is smoother than raw data, but more indicative of trends than weekly average:

```python
start, end = '2017-01', '2017-06'
fig, ax = plt.subplots()
ax.plot(df.loc[start:end, 'solar'],
marker='.', linestyle='-', linewidth=0.5, label='Daily')
ax.plot(dfw.loc[start:end, 'solar'],
marker='o', markersize=8, linestyle='-', label='Weekly Mean Resample')
ax.plot(df7d.loc[start:end, 'solar'],
marker='.', linestyle='-', label='7-d Rolling Mean')
ax.set_ylabel('Solar Production (GWh)')
ax.legend();
```

By increasing the rolling window size, we may emphasize and discover trends over longer periods. 

```python
df365d = df.rolling(window=365, center=True, min_periods=360).mean()
```

Here's a 365-day rolling average, compared to the 7-day rolling average as well as raw daily data. While we could notice the same trends (e.g. lower consumption in 2009 and 2012–2013) in the 7-day rolling average, a 365-day average makes it much more evident:

```python
fig, ax = plt.subplots()
ax.plot(df['consumption'], marker='.', markersize=2, color='0.6',
linestyle='None', label='Daily')
ax.plot(df7d['consumption'], linewidth=2, label='7-d Rolling Mean')
ax.plot(df365d['consumption'], color='0.2', linewidth=3,
label='Trend (365-d Rolling Mean)')
# Set x-ticks to yearly interval and add legend and labels
ax.xaxis.set_major_locator(mdates.YearLocator())
ax.legend()
ax.set_xlabel('Year')
ax.set_ylabel('Consumption (GWh)')
ax.set_title('Trends in Electricity Consumption');
```

We can analyse renewable generation in a similar fashion:

```python
fig, ax = plt.subplots()
for nm in ['wind', 'solar', 'wind_solar']:
    ax.plot(df365d[nm], label=nm)
    # Set x-ticks to yearly interval, adjust y-axis limits, add legend and labels
    ax.xaxis.set_major_locator(mdates.YearLocator())
    ax.set_ylim(0, 400)
    ax.legend()
    ax.set_ylabel('Production (GWh)')
    ax.set_title('Trends in Electricity Production (365-d Rolling Means)');
```

## Aligning time series

In real life, we often encounter sets time series with identical frequencies (e.g. 1 reading/minute), but where the timestamps are not identical. Let us image a set of temperature sensors scattered around the city.

```python
dft = pd.read_csv('temperature.csv')
dft.timestamp = dft.timestamp.apply(pd.to_datetime)
dft.head()
```

```python
len(dft)
```

Please notice that while every sensor produces values every minute, the *seconds* part of the timestamps differs. In most cases, we will be interested in creating a table which allows us to compare the readings from a given point in time. However, a simple pivot will not suffice:

```python
dfp = dft.pivot(index='timestamp', columns='device_id', values='value')
dfp
```

Resampling will handle this quite well though:

```python
dfr = dfp.resample('min').mean()
dfr.head()
```

```python
dfr.plot()
```

However, that will only give us the mean (or otherwise aggregated) value for every time window. 

In the plot above, we can see that one of the sensors must have been off for an entire day (ID *40*). In order to fill any possible holes, we need to use a trick which consists of several steps:

* creating a new `DatetimeIndex` with all the desired time points,
* creating a union of the new index and the original index,
* applying the unified index to the DataFrame,
* interpolating the values,
* removing all index entries other than the desired (regular) ones.

```python
new_index = pd.date_range(dfp.index.round('1min').min(), dfp.index.round('1min').max(), freq='1min')
new_index
```

```python
dfp.reindex(new_index)
```

```python
dfp2 = dfp.reindex(new_index.union(dfp.index)).interpolate()
dfp2.head()
```

```python
dfp3 = dfp2.reindex(new_index)
dfp3.head()
```

```python
dfp3.plot()
```

## Working with time zones

By default, date/time objects are unaware of the time zone (“naive”):

```python
rng = pd.date_range("3/6/2012 00:00", periods=48, freq="h")
rng
```

```python
rng.tz is None
```

An object can be tagged with the timezone upon creation by specifying the `tz` argument:

```python
rng = pd.date_range("3/6/2012 00:00", periods=48, freq="h", tz="Europe/Warsaw")
rng
```

```python
rng.tz
```

An existing object can be localized by calling its `tz_localize` method:

```python
rng = pd.date_range("3/6/2012 00:00", periods=48, freq="h")
rng = rng.tz_localize("Europe/Warsaw")
rng.tz
```

Instead of `pytz` objects, we can also use `dateutil` ones:

```python
rng = pd.date_range("3/6/2012 00:00", periods=48, freq="h")
rng = rng.tz_localize("dateutil/Europe/Warsaw")
rng.tz
```

Let's transform our index (which is in Warsaw's time zone) to London. Note how the times got shifted by an hour:

```python
rng_london = rng.tz_convert("Europe/London")
rng_london
```

Of course, the timestamp values of the $n$-th item in the index will differ (as do their timezones):

```python
rng[3]
```

```python
rng_london[3]
```

However, they will still be treated as equal when comparing:

```python
rng[3] == rng_london[3]
```

Likewise, timestamps with identical *values*, but in different *time zones* are treated as different:

```python
rng_london[4]
```

```python
rng[3] == rng_london[4]
```
